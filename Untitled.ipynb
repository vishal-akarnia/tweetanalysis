```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
from textblob import TextBlob
import re
from wordcloud import WordCloud
import nltk
import string
```

```python
data_train=pd.read_csv('train_2kmZucJ.csv')
data_train.head(20)
```

## Cleaning the Data

```python
## clean the text
def cleanText(text):
    text=re.sub(r'@[A-Za-z0-9]+','', text)
    text=re.sub(r'#','', text)
    text=re.sub(r'RT[\S]+','', text)
    text=re.sub(r'https?:\/\/\+S','', text)
    
    return text


data_train['tweet']=data_train['tweet'].apply(cleanText)
```

```python
#showing the clean data
data_train.drop(['id'],axis=1,inplace=True)
data_train.head()
```

## removing the special charecters

```python
string.punctuation   #need to removie this type of charecter
```

```python
type(data_train['tweet'])    # data type is pandas.core.series i need to convert it in a string
```

```python
data_train['tidey_tweet']=data_train['tweet'].str.replace('[^a-zA-Z#]'," ")
```

```python
data_train.head()
```

## Tokenize the tweet
as we clean the unwanted symboles and patterns we get only words for machine its difficult to differentiate between the words like having or have ,looking or look so we tokentize the words for machine,having converted into have similarly other words too

```python
tokenized_tweet=data_train['tidey_tweet'].apply(lambda x:x.split())
tokenized_tweet.head()
```

```python
## now the stemming part
from nltk.stem.porter import *
stemmer=PorterStemmer()
tokenized_tweet=tokenized_tweet.apply(lambda x:[stemmer.stem(i) for i in x])
tokenized_tweet.head()
```

## Joinning the tokenized word in the same data

```python
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i]=' '.join(tokenized_tweet[i])
data_train['tidey_tweet']=tokenized_tweet

data_train.head()
```

## Adding a column for length of the tweet and punctuation

```python
def count_punct(text):
    count=sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text)-text.count(" ")),3)*100
```

```python
data_train["tweet_len"]=data_train['tweet'].apply(lambda x:len(x)-x.count(" "))
data_train["punct%"]=data_train["tweet"].apply(lambda x:count_punct(x))
data_train.head()
```

## Features selection and Engineering

### 1.count vectorizer
### 2.Tfidf


```python
from sklearn.feature_extraction.text import CountVectorizer

bow_vectorizer=CountVectorizer(stop_words='english')

bow=bow_vectorizer.fit_transform(data_train['tidey_tweet'])
x_count_feat=pd.concat([data_train['tweet_len'],data_train['punct%'],pd.DataFrame(bow.toarray())],axis=1)


x_count_feat.head()
```

```python
# now with the Tfidf

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer=TfidfVectorizer(stop_words='english')
tfidf=tfidf_vectorizer.fit_transform(data_train['tidey_tweet'])
x_tfidf_feat=pd.concat([data_train['tweet_len'],data_train['punct%'],pd.DataFrame(tfidf.toarray())],axis=1)


x_tfidf_feat.head()
```

## Algorithem testing

```python
x=x_count_feat

y=data_train['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y , test_size=0.2 , random_state=21 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_LR= LogisticRegression()

clf_LR.fit(x_train, y_train)

y_pred_LR= clf_LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
print('Accuracy_score: ',accuracy_score(y_test, y_pred_LR)*100)
print('\n','Classification_report: ','\n','\n',classification_report(y_test, y_pred_LR))
```
